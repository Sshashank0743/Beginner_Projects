## Draw A Bar Chart

```python
import matplotlib.pyplot as plt

f = plt.figure(figsize=(10, 10))
plt.matshow(corr_mat, fignum=f.number)
plt.colorbar()
```

```python
from scipy import stats

print(stats.percentileofscore(train["Result"], 106))
```

```python
train.drop(train[train["Result"] >= 106].index, inplace=True)
train.head()
```

```python
train_grouped = train.groupby('key')
target = pd.DataFrame(train_grouped['Result'].mean())
target
```
